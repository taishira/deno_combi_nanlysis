```python
import pandas as pd
import ast
from packaging import version
import re


df_combi = pd.read_csv('std_non_sorted.csv')
combi_li = df_combi['combination'].values.tolist()
version_li = df_combi['rekkyo'].values.tolist()
version_li2 = df_combi['rekkyo'].values.tolist()

li = []
b = [] 
version_li_str = []
version_setted = []

hidari_li = []
kansei_li = []

for i in range(len(combi_li)):
    combi_li[i]=ast.literal_eval(combi_li[i])
    version_li[i]=ast.literal_eval(version_li[i])
    version_li_str.append(version_li[i])
    version_li2[i]=ast.literal_eval(version_li2[i])
    
#print(version_li[0])
print('A')
for i in range(len(version_li_str)):
    for j in range(len(version_li_str[i])):
        version_li_str[i][j] = str(version_li_str[i][j])

for i in range(len(version_li_str)):
    version_li_str[i] = list(set(version_li_str[i]))
    version_setted.append([])
    for j in version_li_str[i]:
        version_setted[i].append(ast.literal_eval(j))
        
print(version_setted[0])

#######

for i in range(len(version_setted)):
    hidari_li.append([])
    for j in range(len(version_setted[i])):
        hidari_li[i].append(version_setted[i][j][0])

        
print(hidari_li[1])
        
for i in range(len(hidari_li)):
    for j in range(0,len(hidari_li[i])-1,1):
        for k in range(j,len(hidari_li[i]),1):
            if version.parse(hidari_li[i][k]) < version.parse(hidari_li[i][j]):
                minimal_ver = hidari_li[i][j]
                hidari_li[i][j] = hidari_li[i][k]
                hidari_li[i][k] = minimal_ver
    
    



for i in range(len(hidari_li)):
    kansei_li.append([])
    for j in range(len(hidari_li[i])):
        minimal_li = []
        for k in range(len(version_setted[i])):
            if k not in minimal_li:
                if hidari_li[i][j] == version_setted[i][k][0]:
                    kansei_li[i].append(version_setted[i][k])

#########                    
print(kansei_li[1])       

ver_joui_li = []
for i in range(len(kansei_li)):
    ichibu = 0
    kanzen = 0
    onaji= 0
    sougo = 0
    sougo_ichibu = 0
    vernashi = 0
    #print(i)
    if len(kansei_li[i]) == 1:
        onaji += (len(version_li[i]) - len(kansei_li[i]))
    elif len(kansei_li[i]) > 1:
        
        for j in range(len(kansei_li[i])-1):
            for k in range(j+1,len(kansei_li[i]),1):
                version_hikaku = [0,0,0,0]

                for l in range(len(kansei_li[i][j])):
                    if ('0.0.0' in kansei_li[i][j]) or ('0.0.0' in kansei_li[i][k]):
                        version_hikaku[3] += 1
                    if version.parse(kansei_li[i][j][l]) < version.parse(kansei_li[i][k][l]):
                        version_hikaku[0] += 1
                    if version.parse(kansei_li[i][j][l]) == version.parse(kansei_li[i][k][l]):
                        version_hikaku[1] += 1
                    if version.parse(kansei_li[i][j][l]) > version.parse(kansei_li[i][k][l]):
                        version_hikaku[2] += 1
                if ((version_hikaku[0] > 0) ^ (version_hikaku[2] > 0)) and version_hikaku[1] == 0 and version_hikaku[3] == 0:
                    kanzen += 1
                elif version_hikaku[1] > 0  and ((version_hikaku[0] > 0) ^ (version_hikaku[2] > 0)) and version_hikaku[3] == 0:
                    ichibu += 1
                elif version_hikaku[0] > 0 and version_hikaku[1] == 0 and version_hikaku[2] > 0 and version_hikaku[3] == 0:
                    sougo += 1
                elif version_hikaku[0] > 0 and version_hikaku[1] > 0 and version_hikaku[2] > 0 and version_hikaku[3] == 0:
                    sougo_ichibu += 1
                elif version_hikaku[3] > 0:
                    vernashi += 1
    onaji = len(version_li) - (kanzen + ichibu + sougo + sougo_ichibu + vernashi)
    ver_joui_li.append([kanzen,ichibu,sougo,sougo_ichibu,onaji,vernashi,len(kansei_li[i]),len(version_li[i])])
    
    
#vernashi_count_li = []
#for i in range(len(version_li2)):
#    vernashi_count = 0
#    for l in range(len(version_li2[i])):
#        if '0.0.0' in version_li2[i][l]:
#            vernashi_count+=1
#    ver_joui_li[i].append(vernashi_count)
    
    
#print(vernashi_count_li)

for i in range(len(version_li)):
    b.append([])
    b[i].append(combi_li[i])
    b[i].append(version_li2[i])
    b[i].append(ver_joui_li[i])
    
k1 = pd.DataFrame(b,
                  columns=['combination','rekkyo','joui'])
k1.to_csv("std_non_joui2.csv", encoding="shift_jis", index = False)
```
